```python
df1['ParentEduc'].value_counts()
```

```python
prntEduGp = df1.groupby('ParentEduc').agg({'MathScore': 'mean', 'ReadingScore': 'mean', 'WritingScore': 'mean'})
prntEduGp
```

```python
sns.heatmap(prntEduGp, cmap= 'YlGnBu', annot= True)
plt.title('Parrents Education Impact on Students Score')
```

```python
df1['ParentMaritalStatus'].unique()
```

```python
prntMrrdGp = df1.groupby('ParentMaritalStatus').agg({'MathScore': 'mean', 'ReadingScore': 'mean', 'WritingScore': 'mean'})
prntMrrdGp
```

```python
sns.heatmap(prntMrrdGp, cmap= 'YlGnBu', annot= True)
plt.title('Parrents Merrital Impact on Students Score')
plt.ylabel('Parrents Merrital Status')
plt.show()
```
